```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('../dados/chamados_com_predicoes.csv', index_col='id_chamado')
```

# 1 - EDA

```python
dados.dtypes
```

```python
dados['data_abertura'] = pd.to_datetime(dados['data_abertura'])
```

```python
dados.isna().any()
```

```python
dados['bairro'].value_counts(normalize=True)
```

```python
dados['categoria_real'].value_counts(normalize=True)
```

```python
mapeamento_zonas = {
    'Campo Grande': 'Zona Oeste',
    'Bangu': 'Zona Oeste',
    'Santa Cruz': 'Zona Oeste',
    'Realengo': 'Zona Oeste',
    
    'Tijuca': 'Zona Norte',
    'Madureira': 'Zona Norte',
    'Irajá': 'Zona Norte',
    'Méier': 'Zona Norte',
    'Penha': 'Zona Norte',
    'Ilha do Governador': 'Zona Norte',
    
    'Jacarepaguá': 'Zona Sudoeste',
    'Barra da Tijuca': 'Zona Sudoeste',
    
    'Copacabana': 'Zona Sul',
    'Botafogo': 'Zona Sul',
    
    'Centro': 'Centro'
}
```

```python
dados['zona'] = dados['bairro'].map(mapeamento_zonas)
```

```python
dados.head()
```

```python
#Gerar gráfico de frequência global de ocorrências
plt.figure(figsize=(8, 5))
frequencia_categorias = dados['categoria_real'].value_counts()
frequencia_categorias.plot(kind='bar', color='royalblue', edgecolor='black', width=0.6)
plt.title(f'Frequência de Chamados por Categoria', fontsize=12, fontweight='bold', pad=12)
plt.xlabel('Categoria Real', fontsize=10)
plt.ylabel('Quantidade de Chamados', fontsize=10)
plt.xticks(rotation=25, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.plot()
nome_arquivo = f"../results/figures/grafico_frequencia_ocorrencias_global.png"
plt.savefig(nome_arquivo, dpi=300)
plt.close()
```

```python
# Gerar gráficos de frequência de ocorrências por bairro
for bairro, dados_bairro in dados.groupby('bairro'):
    frequencia_categorias = dados_bairro['categoria_real'].value_counts()
    plt.figure(figsize=(8, 5))
    frequencia_categorias.plot(kind='bar', color='royalblue', edgecolor='black', width=0.6)
    plt.title(f'Frequência de Chamados - Bairro: {bairro}', fontsize=12, fontweight='bold', pad=12)
    plt.xlabel('Categoria Real', fontsize=10)
    plt.ylabel('Quantidade de Chamados', fontsize=10)
    plt.xticks(rotation=25, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    nome_arquivo = f"../results/figures/grafico_frequencia_ocorrencias_{bairro.lower().replace(' ', '_')}.png"
    plt.savefig(nome_arquivo, dpi=300)
    plt.close()
```

```python
#Gerar gráfico de frequência de canal de chamado
plt.figure(figsize=(8, 5))
frequencia_canal = dados['canal'].value_counts()
frequencia_canal.plot(kind='bar', color='royalblue', edgecolor='black', width=0.6)
plt.title(f'Frequência de Chamados por Canal', fontsize=12, fontweight='bold', pad=12)
plt.xlabel('Canal', fontsize=10)
plt.ylabel('Quantidade de Chamados', fontsize=10)
plt.xticks(rotation=25, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.plot()
nome_arquivo = f"../results/figures/grafico_frequencia_canal_global.png"
plt.savefig(nome_arquivo, dpi=300)
plt.close()
```

```python
bimestres = [
    {"nome": "1º Bimestre (Jan-Fev)", "inicio": "2026-01-01", "fim": "2026-02-28"},
    {"nome": "2º Bimestre (Mar-Abr)", "inicio": "2026-03-01", "fim": "2026-04-30"},
    {"nome": "3º Bimestre (Mai-Jun)", "inicio": "2026-05-01", "fim": "2026-06-30"}
]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6), sharey=True)
ordem_categorias = dados['categoria_real'].value_counts().index

for i, bim in enumerate(bimestres):
    ax = axes[i]
    dados_filtrados = dados[(dados['data_abertura'] >= bim["inicio"]) & (dados['data_abertura'] <= bim["fim"])]
    frequencia_categorias = dados_filtrados['categoria_real'].value_counts().reindex(ordem_categorias, fill_value=0)
    frequencia_categorias.plot(kind='bar', ax=ax, color='royalblue', edgecolor='black', width=0.6)
    ax.set_title(bim["nome"], fontsize=12, fontweight='bold', pad=10)
    ax.set_xlabel('Categoria Real', fontsize=10)
    ax.set_xticklabels(ordem_categorias, rotation=35, ha='right', fontsize=9)
    ax.grid(axis='y', linestyle='--', alpha=0.6)
    if i == 0:
        ax.set_ylabel('Quantidade de Chamados', fontsize=10)

plt.suptitle('Frequência de Chamados por Categoria - Primeiro Semestre de 2026', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
nome_arquivo = "../results/figures/grafico_frequencia_ocorrencias_bimestral.png"
plt.savefig(nome_arquivo, dpi=300, bbox_inches='tight')
plt.close()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6), sharey=True)
ordem_canais = dados['canal'].value_counts().index

for i, bim in enumerate(bimestres):
    ax = axes[i]
    dados_filtrados = dados[(dados['data_abertura'] >= bim["inicio"]) & (dados['data_abertura'] <= bim["fim"])]
    frequencia_canal = dados_filtrados['canal'].value_counts().reindex(ordem_canais, fill_value=0)
    frequencia_canal.plot(kind='bar', ax=ax, color='royalblue', edgecolor='black', width=0.6)
    ax.set_title(bim["nome"], fontsize=12, fontweight='bold', pad=10)
    ax.set_xlabel('Canal', fontsize=10)
    ax.set_xticklabels(ordem_canais, rotation=35, ha='right', fontsize=9)
    ax.grid(axis='y', linestyle='--', alpha=0.6)
    if i == 0:
        ax.set_ylabel('Quantidade de Chamados', fontsize=10)

plt.suptitle('Frequência de Chamados por Canal - Primeiro Semestre de 2026', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
nome_arquivo = "../results/figures/grafico_frequencia_canal_bimestral.png"
plt.savefig(nome_arquivo, dpi=300, bbox_inches='tight')
plt.close()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12), sharey=True)
ordem_categorias = dados['categoria_real'].value_counts().index
i = 0

for zona, dados_zona in dados.groupby('zona'):
    ax = axes[i//3, i%3]
    frequencia_categorias = dados_zona['categoria_real'].value_counts().reindex(ordem_categorias, fill_value=0)
    frequencia_categorias.plot(kind='bar', ax=ax, color='royalblue', edgecolor='black', width=0.6)
    ax.set_title(zona, fontsize=12, fontweight='bold', pad=10)
    ax.set_xlabel('Categoria Real', fontsize=10)
    ax.set_xticklabels(ordem_categorias, rotation=35, ha='right', fontsize=9)
    ax.grid(axis='y', linestyle='--', alpha=0.6)
    i+=1
    if i%3 == 0:
        ax.set_ylabel('Quantidade de Chamados', fontsize=10)

plt.suptitle('Frequência de Chamados por Zona', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
nome_arquivo = "../results/figures/grafico_frequencia_ocorrencias_bimestral_zona.png"
plt.savefig(nome_arquivo, dpi=300, bbox_inches='tight')
plt.close()
```

```python
ordem_categorias = dados['categoria_real'].value_counts().index


# Gerar gráficos de frequência de ocorrências por bairro
for bairro, dados_bairro in dados.groupby('bairro'):
    
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6), sharey=True)    
    
    for i, bim in enumerate(bimestres):
        ax = axes[i]
        dados_filtrados = dados_bairro[(dados_bairro['data_abertura'] >= bim["inicio"]) & (dados_bairro['data_abertura'] <= bim["fim"])]
        frequencia_categorias = dados_filtrados['categoria_real'].value_counts().reindex(ordem_categorias, fill_value=0)
        frequencia_categorias.plot(kind='bar', ax=ax, color='royalblue', edgecolor='black', width=0.6)
        
        ax.set_title(f'{bim["nome"]}', fontsize=12, fontweight='bold', pad=12)
        ax.set_xlabel('Categoria Real', fontsize=10)
        if i==0:
            ax.set_ylabel('Quantidade de Chamados', fontsize=10)
        ax.set_xticklabels(ordem_categorias, rotation=25, ha='right')
        ax.grid(axis='y', linestyle='--', alpha=0.6)
    
    plt.suptitle(f'Evolução de Chamados - Bairro: {bairro}', fontsize=14, fontweight='bold', y=1.02)
    plt.tight_layout()
    nome_arquivo = f"../results/figures/bairros_bimestral/grafico_frequencia_ocorrencias_{bairro.lower().replace(' ', '_')}_bimestral.png"
    plt.savefig(nome_arquivo, dpi=300, bbox_inches='tight')
    plt.close()
```

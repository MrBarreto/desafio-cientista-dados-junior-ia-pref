```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('../dados/chamados_com_predicoes.csv', index_col='id_chamado')
```

```python
dados.dtypes
```

```python
dados['data_abertura'] = pd.to_datetime(dados['data_abertura'])
```

```python
dados.isna().any()
```

```python
dados['bairro'].value_counts(normalize=True)
```

```python
dados['categoria_real'].value_counts(normalize=True)
```

```python
mapeamento_zonas = {
    'Campo Grande': 'Zona Oeste',
    'Bangu': 'Zona Oeste',
    'Santa Cruz': 'Zona Oeste',
    'Realengo': 'Zona Oeste',
    
    'Tijuca': 'Zona Norte',
    'Madureira': 'Zona Norte',
    'Irajá': 'Zona Norte',
    'Méier': 'Zona Norte',
    'Penha': 'Zona Norte',
    'Ilha do Governador': 'Zona Norte',
    
    'Jacarepaguá': 'Zona Sudoeste',
    'Barra da Tijuca': 'Zona Sudoeste',
    
    'Copacabana': 'Zona Sul',
    'Botafogo': 'Zona Sul',
    
    'Centro': 'Centro'
}
```

```python
dados['zona'] = dados['bairro'].map(mapeamento_zonas)
```

```python
dados.head()
```

```python
#Gerar gráfico de frequência global de ocorrências
plt.figure(figsize=(8, 5))
frequencia_categorias = dados['categoria_real'].value_counts()
frequencia_categorias.plot(kind='bar', color='royalblue', edgecolor='black', width=0.6)
plt.title(f'Frequência de Chamados por Categoria', fontsize=12, fontweight='bold', pad=12)
plt.xlabel('Categoria Real', fontsize=10)
plt.ylabel('Quantidade de Chamados', fontsize=10)
plt.xticks(rotation=25, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.plot()
nome_arquivo = f"../results/figures/grafico_frequencia_ocorrencias_global.png"
plt.savefig(nome_arquivo, dpi=300)
plt.close()
```

```python
# Gerar gráficos de frequência de ocorrências por bairro
for bairro, dados_bairro in dados.groupby('bairro'):
    frequencia_categorias = dados_bairro['categoria_real'].value_counts()
    plt.figure(figsize=(8, 5))
    frequencia_categorias.plot(kind='bar', color='royalblue', edgecolor='black', width=0.6)
    plt.title(f'Frequência de Chamados - Bairro: {bairro}', fontsize=12, fontweight='bold', pad=12)
    plt.xlabel('Categoria Real', fontsize=10)
    plt.ylabel('Quantidade de Chamados', fontsize=10)
    plt.xticks(rotation=25, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    nome_arquivo = f"../results/figures/grafico_frequencia_ocorrencias_{bairro.lower().replace(' ', '_')}.png"
    plt.savefig(nome_arquivo, dpi=300)
    plt.close()
```
